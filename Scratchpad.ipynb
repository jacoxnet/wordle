```python
def createSortForm(recordInstance):
    form = []
    for lDict in recordInstance.values():
        form.append(list(lDict.keys())[0])
        form.append(list(lDict.values())[0])
    return form
```

```python

x = [createSortForm(record['cigar']), createSortForm(record['sissy'])]
print(sorted(x))
```

```python
z = []
for rec in record.values():
    z.append(createSortForm(rec))
zp = sorted(z, key=lambda k: (''.join(k)))
zp
```
